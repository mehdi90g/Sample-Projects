In this task, we take a raw dataset containing the information on the number of different vehicle types in various cities, along with the population of each city (the numbers of the input dataset are randomly generated). The goal is to automatically transform the raw dataset into a new and much more concise dataset where the we have the information on the percentage of population in each city which own a particular vehicle type.

Loading the input dataset into a pandas DataFrame structure.

```python
Input_Data = pd.read_csv('Raw_ETL_Dataset.csv', sep=',')

display(Input_Data.head())
```

```python
Percentage = [[] for _ in range(Input_Data.shape[0])]
for i in range(Input_Data.shape[0]):
  Percentage[i] = (Input_Data['Number_of_Vehicles'][i]/Input_Data['Total_Population'][i])*100
print(Percentage)
```

```python
Input_Data['Percentage'] = Percentage
display(Input_Data.head())
```

```python
a = Input_Data['City_ID'].unique()
a = sorted(a)
#print(a)
b = Input_Data['Vehicle_Type'].unique()
b = sorted(b)
#print(b)
b2 = ['City_ID'] + b
#print(b2)
Output_Data = {}
for i in b2:
  Output_Data[i] = []
Output_Data =  pd.DataFrame(Output_Data)
Output_Data['City_ID'] = a
print(Output_Data)
```

```python
for i in range(len(a)):
  for j in range(len(b)):
    filterDataframe = Input_Data[(Input_Data['City_ID'] == a[i]) & (Input_Data['Vehicle_Type'] == b[j]) ]
    ind=filterDataframe['Percentage'].index
    ind=ind.tolist()
    if ind:   #i.e., if ind is a nonempty list
      ind=ind[0]
      Output_Data.iat[i,j+1] = round(filterDataframe['Percentage'][ind], 2)
    else:   #i.e., if ind is an empty list
      Output_Data.iat[i,j+1] = round(0, 0)
print(Output_Data)
```

```python
Output_Data.to_csv("Transformed_ETL_Dataset.csv", index=False)

files.download('Transformed_ETL_Dataset.csv')
```
